```python
text =  open('drive/MyDrive/AI/PJ3/weight/pianoabc.txt','r').read()

uniqueText = list(set(text)) # 중복을 제거하는 list
uniqueText.sort()

# utilities
text_to_num = {}
num_to_text = {}

for idx, data in enumerate(uniqueText):
  text_to_num[data] = idx
  num_to_text[idx]=data

digitizationText = []

for i in text:
  digitizationText.append(text_to_num[i])

X = []
Y = []

for i in range(0,len(digitizationText)-25):
  X.append(digitizationText[ i : i+25 ])
  Y.append(digitizationText[i + 25 ])

import numpy as np
import tensorflow as tf

print(np.array(X).shape)

X = tf.one_hot(X,31)
Y = tf.one_hot(Y,31)
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(100,input_shape=(25,31) ),
    tf.keras.layers.Dense(31,activation='softmax')
])

model.compile(loss='categorical_crossentropy',optimizer='adam')
model.fit( X, Y, batch_size=64, epochs=50, verbose=2 )

model.save('drive/MyDrive/AI/PJ3/weight')
```

```python
import tensorflow as tf

Pmodel = tf.keras.models.load_model('drive/MyDrive/AI/PJ3/weight')

text =  open('drive/MyDrive/AI/PJ3/weight/pianoabc.txt','r').read()

uniqueText = list(set(text)) # 중복을 제거하는 list
uniqueText.sort()

# utilities
text_to_num = {}
num_to_text = {}

for idx, data in enumerate(uniqueText):
  text_to_num[data] = idx
  num_to_text[idx]=data

digitizationText = []

for i in text:
  digitizationText.append(text_to_num[i])

import numpy as np

firstInput = digitizationText[117:117+25]
firstInput = tf.one_hot(firstInput, 31)
firstInput = tf.expand_dims(firstInput,axis=0)

predictValue = Pmodel.predict(firstInput)
predictValue = np.argmax(predictValue[0])
print(num_to_text[predictValue])
```

```python
import tensorflow as tf

music = []

for i in range(200) :
  predictValue = Pmodel.predict(firstInput)
  if(i%12==0):
    new_value = np.random.choice(uniqueText,1,p=predictValue[0]) #랜덤성
    new_value = text_to_num[new_value[0]]
    predictValue = new_value
  else:
    predictValue = np.argmax(predictValue[0])

  music.append(predictValue)

  nextInput = firstInput.numpy()[0][1:]
  one_hot_num = tf.one_hot(predictValue,31)

  firstInput = np.vstack([ nextInput, one_hot_num.numpy() ])
  firstInput = tf.expand_dims(firstInput,axis=0)


print(music)

```

```python
music_text = []

for i in music:
  music_text.append(num_to_text[i])

print(''.join(music_text))
```
